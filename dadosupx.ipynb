```python
import matplotlib.pyplot as plt


coluna_destino = df['Destination'].value_counts().head(10)


plt.figure(figsize=(10, 6))
coluna_destino.plot(kind='bar', color='purple')
plt.title('Top 10 Destinos Mais Visitados')
plt.xlabel('Destino')
plt.ylabel('Número de Visitas')
plt.xticks(rotation=45)  
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


destinos_por_genero = df.groupby(['Destination', 'Traveler gender']).size().reset_index(name='VisitCount')


top_destinos = df['Destination'].value_counts().head(10).index
destinos_por_genero = destinos_por_genero[destinos_por_genero['Destination'].isin(top_destinos)]


plt.figure(figsize=(12, 8))
sns.barplot(data=destinos_por_genero, x='Destination', y='VisitCount', hue='Traveler gender')
plt.title('Top 10 Destinos Mais Visitados por Gênero')
plt.xlabel('Destino')
plt.ylabel('Número de Visitas')
plt.xticks(rotation=45)
plt.legend(title='Gênero')
plt.show()
```
